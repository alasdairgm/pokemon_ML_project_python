# Pokemon data analysis

**Context** (taken from Kaggle)

This dataset contains information on all 802 Pokemon from all Seven Generations of Pokemon. The information contained in this dataset include Base Stats, Performance against Other Types, Height, Weight, Classification, Egg Steps, Experience Points, Abilities, etc. The information was scraped from http://serebii.net/

**Content**

- name: The English name of the Pokemon
- japanese_name: The Original Japanese name of the Pokemon
- pokedex_number: The entry number of the Pokemon in the National Pokedex
- percentage_male: The percentage of the species that are male. Blank if the Pokemon is genderless.
- type1: The Primary Type of the Pokemon
- type2: The Secondary Type of the Pokemon
- classification: The Classification of the Pokemon as described by the Sun and Moon Pokedex
- height_m: Height of the Pokemon in metres
- weight_kg: The Weight of the Pokemon in kilograms
- capture_rate: Capture Rate of the Pokemon
- base_egg_steps: The number of steps required to hatch an egg of the Pokemon
- abilities: A stringified list of abilities that the Pokemon is capable of having
- experience_growth: The Experience Growth of the Pokemon
- base_happiness: Base Happiness of the Pokemon
- against_?: Eighteen features that denote the amount of damage taken against an attack of a particular type
- hp: The Base HP of the Pokemon
- attack: The Base Attack of the Pokemon
- defense: The Base Defense of the Pokemon
- sp_attack: The Base Special Attack of the Pokemon
- sp_defense: The Base Special Defense of the Pokemon
- speed: The Base Speed of the Pokemon
- generation: The numbered generation which the Pokemon was first introduced
- is_legendary: Denotes if the Pokemon is legendary.

```python
import statsmodels.api as sm

# add in the constant to the data
predictor_vars = sm.add_constant(predictor_vars)
predictor_vars.head()
```

```python
sm_model = sm.OLS(response_var, predictor_vars).fit()
print(sm_model.summary())
```

## Diagnostic plots

```python
#first calculate the residuals
residuals = sm_model.fittedvalues - response_var
```

### Residual vs fitted plot

```python
%matplotlib inline
import seaborn as sns 
import matplotlib.pyplot as plt

sns.residplot(x = sm_model.fittedvalues, y = residuals, # data to plot
                          lowess = True, # Fit a lowess smoother to the residual scatterplot
                          scatter_kws = {'alpha': 0.5}, # set the transparency of the scatter points
                          line_kws = {'color': 'red', 'lw': 1, 'alpha': 0.8}) # set the line colour/transparency
plt.xlabel("Fitted value")
plt.ylabel("Residual")
```

### Normal QQ plot

```python
import scipy.stats as stats

sm.qqplot(residuals, dist=stats.t, fit=True, line='45')
```

## Logistic regression

Let's see if I can predict whether a pokemon is legendary.

```python
response_var = pokemon["is_legendary"]
predictor_vars = pokemon.drop(columns = "is_legendary")
```

```python
from sklearn.model_selection import train_test_split

poke_pred_train, poke_pred_test, poke_resp_train, poke_resp_test = (
    train_test_split(
        predictor_vars, 
        response_var, 
        test_size = 0.2,  # test set is 20%
        random_state = 7
    )
)
```

```python
response_var
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

model = LogisticRegression()
model.fit(poke_pred_train, poke_resp_train)
scores = cross_val_score(
    model, poke_pred_train, poke_resp_train, scoring='accuracy', cv=10
)
scores
```

Now we can see the mean accuracy on the training data. Like with LinearRegression we know the metric returned from score() is accuracy

```python
print(model.score(poke_pred_train, poke_resp_train))
print(np.mean(scores))
```

And on the test data...

```python
model.score(poke_pred_test, poke_resp_test)
```

We can also get other metrics such as ROC and AUC. Recalling that to calculate the ROC what we want is the predicted probabilities and then the observed values in the data. First we want to return the predicted probabilities from our model:

```python
pred_test = model.predict_proba(poke_pred_test)
pred_test
```

We see we get an array with 2 entries for every row of the data - these are the probabilities for both outcomes (0 or 1) for the paid column - i.e. every row will add to 1. We just want the probabilities of the positive outcome - that the pokemon is legendary - which is the right hand column of the array:



```python
pred_test = pred_test[:, 1]
pred_test
```

Now we have the 2 pieces of information required to calculate our AUC score from the ROC. We use the roc_auc_score function to calculate this:

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(y_true = poke_resp_test, y_score = pred_test)
```
